```python
#FILTERING DATA FROM DICTIONARIES
import pandas as pd
data={
    "Name":['John','Alice','Jay','Mark'],
    "Age":[20,21,22,19],
    "Marks":[70,85,75,90]
}
df=pd.DataFrame(data)
print("Original data:")
print(df)
#Filtering data
print("Alice's data:")
print(df[df["Name"]=='Alice'])
print("Students with marks less than 85:")
print(df[df["Marks"]<85])
print("Students with marks greater than 75 and age less than 20:")
print(df[(df["Age"]>20)&(df["Marks"]>75)])
```

```python
#DATA MANIPULATION OPERATIONS
import pandas as pd
data={
    'Name':['A','B','C'],
    'Age':[20,21,22],
    'Marks':[70,80,90]
}
df=pd.DataFrame(data)
print("Original data:")
print(df)
#Adding a new column
df['Bonus']=5
print("After adding bonus:")
print(df)
#increasing marks
df['Marks']=df['Marks']+2
print("\nAfter increasing marks:")
print(df)
#create new column from existing columns
df["Total"]=df["Marks"]+df["Bonus"]
print("\nAfter adding total:")
print(df)
#change values
df.loc[1,"Age"]=23
print("\nAfter changing age:")
print(df)
#renaming column
df.rename(columns={"Total":"Final"},inplace=True)
print("\nAfter renaming:")
print(df)
#removing a column
df.drop("Bonus",axis=1,inplace=True)
print("\nAfter removing bonus:")
print(df)
```
