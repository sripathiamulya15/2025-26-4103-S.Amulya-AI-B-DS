```python
# LIST OPERATIONS

# Creating a list
lst = [10, 20, 30, 40]
print("Initial list:", lst)

# Accessing elements
print("First element:", lst[0])
print("Last element:", lst[-1])

# Adding elements
lst.append(50)
print("After append:", lst)

lst.insert(2, 25)
print("After insert:", lst)

lst.extend([60, 70])
print("After extend:", lst)

# Removing elements
lst.remove(25)
print("After remove:", lst)

lst.pop()
print("After pop:", lst)

lst.pop(1)
print("After pop at index 1:", lst)

# Updating element
lst[0] = 100
print("After updating:", lst)

# Searching
print("Index of 30:", lst.index(30))
print("Count of 20:", lst.count(20))

# Sorting
lst.sort()
print("After sort:", lst)

lst.sort(reverse=True)
print("After reverse sort:", lst)

# Reversing
lst.reverse()
print("After reverse:", lst)

# Copying
new_lst = lst.copy()
print("Copied list:", new_lst)

# Length
print("Length of list:", len(lst))

# Membership
print("Is 40 in list?", 40 in lst)

# Looping
print("Elements in list:")
for item in lst:
    print(item)

# List comprehension
squares = [x*x for x in lst]
print("Squares of elements:", squares)

# Clearing list
lst.clear()
print("After clear:", lst)
```

```python
# TUPLE OPERATIONS

# Creating tuples
t = (10, 20, 30, 40, 20)
print("Original tuple:", t)

# Accessing elements
print("First element:", t[0])
print("Last element:", t[-1])

# Slicing
print("Sliced tuple:", t[1:4])

# Length
print("Length of tuple:", len(t))

# Membership
print("Is 30 in tuple?", 30 in t)

# Counting elements
print("Count of 20:", t.count(20))

# Finding index
print("Index of 40:", t.index(40))

# Looping
print("Tuple elements:")
for item in t:
    print(item)

# Concatenation
t2 = (50, 60)
t3 = t + t2
print("After concatenation:", t3)

# Repetition
t4 = t * 2
print("After repetition:", t4)

# Nested tuple
nested = (1, (2, 3), (4, 5))
print("Nested tuple element:", nested[1][0])

# Tuple unpacking
a, b, c, d, e = t
print("Unpacked values:", a, b, c, d, e)

# Converting tuple to list (to modify)
lst = list(t)
lst.append(100)
t_new = tuple(lst)
print("Modified tuple (via list):", t_new)
```

```python
# DICTIONARY OPERATIONS

# Creating a dictionary
d = {
    "name": "Amulya",
    "age": 20,
    "course": "BTech"
}
print("Original dictionary:", d)

# Accessing values
print("Name:", d["name"])
print("Age:", d.get("age"))

# Adding a new key-value pair
d["college"] = "XYZ University"
print("After adding:", d)

# Updating value
d["age"] = 21
print("After updating:", d)

# Removing elements
d.pop("course")
print("After pop:", d)

removed_item = d.popitem()
print("After popitem:", d)
print("Removed item:", removed_item)

# Keys, values, items
print("Keys:", d.keys())
print("Values:", d.values())
print("Items:", d.items())

# Membership
print("Is 'name' a key?", "name" in d)

# Length
print("Length of dictionary:", len(d))

# Looping through dictionary
print("Looping through keys:")
for key in d:
    print(key, ":", d[key])

print("Looping through key-value pairs:")
for k, v in d.items():
    print(k, "->", v)

# Copying dictionary
d_copy = d.copy()
print("Copied dictionary:", d_copy)

# Updating using update()
d.update({"city": "Hyderabad", "age": 22})
print("After update():", d)

# Nested dictionary
student = {
    "id": 1,
    "info": {
        "name": "Amit",
        "marks": 90
    }
}
print("Nested dictionary value:", student["info"]["marks"])

# Clearing dictionary
d.clear()
print("After clear:", d)
```

```python
# SET OPERATIONS
s1 = {10, 20, 30, 40}
s2 = {30, 40, 50, 60}
print("Union using method:", s1.union(s2))
print("Intersection using method:", s1.intersection(s2))
print("Difference using method:", s1.difference(s2))
print("Symmetric difference using method:", s1.symmetric_difference(s2))
```
