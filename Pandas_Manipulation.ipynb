# *GROUPING, REARRANGING AND RANKING DATA*

```python
#GROUPING
import pandas as pd
data={
    "Department":['IT','IT','HR','HR','Finance'],
    "Name":['A','B','C','D','E'],
    "Marks":[80,85,75,90,88]
}
df=pd.DataFrame(data)
print(df)
#mean()
print("Average marks by department:")
print(df.groupby("Department")["Marks"].mean())
#sum()
print("Total marks by department:")
print(df.groupby("Department")["Marks"].sum())
#count()
print("Number of students by department:")
print(df.groupby("Department")["Marks"].count())
```

```python
#REARRANGING
data={
    'Name':['A','B','C'],
    'Age':[19,18,22],
    'Marks':[85,90,80]
}
df=pd.DataFrame(data)
print(df)
#marks arrangement-ascending
df_sorted=df.sort_values(by='Marks')
print("After rearranging marks by ascending order:")
print(df_sorted)
#marks arrangement-descending
df_sorted_desc=df.sort_values(by='Marks',ascending=False)
print("After rearranging marks by descending order:")
print(df_sorted_desc)
#rearranging columns
df_rearranged=df[['Age','Name','Marks']]
print("After rearranging columns:")
print(df_rearranged)
```

```python
#RANKING
import pandas as pd
data={
    'Name':['A','B','C','D'],
    'Marks':[85,92,78,92]
}
df=pd.DataFrame(data)
print(df)
df["Rank"]=df["Marks"].rank(ascending=False)
print("\nAfter ranking:")
print(df)
```
