# *PLOTTING GRAPHS*

### ***DATE : 10/2/2026***

*WEEK 2*



```python
#LINE GRAPH
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("x-y graph")
plt.show()
```

```python
#BAR GRAPH
import matplotlib.pyplot as plt
subjects=['Maths','Physics','Chemistry','CS']
marks=[90,80,70,60]
plt.bar(subjects,marks)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Marks in subjects")
plt.show()
```

### ***DATE : 17/2/2026***

```python
#SCATTERED GRAPH
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[5,7,8,6,9]
plt.scatter(x,y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter graph")
plt.show()
```

```python
#HISTOGRAM
import matplotlib.pyplot as plt
marks=[90,80,70,60,50,40,30,20,10]
plt.hist(marks)
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Histogram")
```

```python
#PIE CHART
import matplotlib.pyplot as plt
activities=['study','sleep','play','others']
hours=[6,4,8,6]
plt.pie(hours,labels=activities,autopct="%1.1f%%")
plt.title("Daily activities")
plt.show()
```
