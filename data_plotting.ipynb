# *PLOTTING GRAPHS*

```python
#LINE GRAPH
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("x-y graph")
plt.show()
```

```python
import matplotlib.pyplot as plt
subjects=['Maths','Physics','Chemistry','CS']
marks=[90,80,70,60]
plt.bar(subjects,marks)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Marks in subjects")
plt.show()
```
